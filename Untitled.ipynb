## Problem 3 
We flip the usual direction of linear regression.

---

### üîπ 1. Generative Model

Instead of predicting `y` from `x`, we now predict `x` from `y`.

**Model**:  
x = œï‚ÇÄ + œï‚ÇÅ¬∑y

This is the generative form.

---

### üîπ 2. New Loss Function

Since we're predicting `x` from `y`, the loss becomes:

L(œï‚ÇÄ, œï‚ÇÅ) = ‚àë (x·µ¢ ‚àí (œï‚ÇÄ + œï‚ÇÅ¬∑y·µ¢))¬≤

This is the same structure as ordinary least squares, but with inputs and outputs swapped.

---

### üîπ 3. Inverse Function

To perform inference (i.e., recover `y` from a new `x`), we invert the function:

x = œï‚ÇÄ + œï‚ÇÅ¬∑y  
‚áí y = (x ‚àí œï‚ÇÄ) / œï‚ÇÅ

So the inverse function is:

y = g‚Åª¬π(x, œï) = (x ‚àí œï‚ÇÄ) / œï‚ÇÅ

---

### üîπ 4. Will the Predictions Match the Discriminative Model?

**No**, in general they will not.

Even though the equations look symmetric, the **error is minimized in different directions**:

- Discriminative regression (y = œï‚ÇÄ + œï‚ÇÅ¬∑x) minimizes **vertical** error (difference in y).
- Generative regression (x = œï‚ÇÄ + œï‚ÇÅ¬∑y) minimizes **horizontal** error (difference in x).

These only match if the data lies exactly on a line (i.e., perfect correlation).

---

### üîπ 5. Python Example: Compare Both Methods



```python
import numpy as np
import matplotlib.pyplot as plt

# Example data
x = np.array([1, 2, 3])
y = np.array([2, 2.5, 4])

# Discriminative: y = a + b*x
A = np.vstack([x, np.ones(len(x))]).T
b_discriminative, a_discriminative = np.linalg.lstsq(A, y, rcond=None)[0]

# Generative: x = a + b*y ‚áí solve x ~ y, then invert
A_gen = np.vstack([y, np.ones(len(y))]).T
b_gen, a_gen = np.linalg.lstsq(A_gen, x, rcond=None)[0]

# Invert to get y = (x - a_gen) / b_gen
b_inverted = 1 / b_gen
a_inverted = -a_gen / b_gen

# Compare lines
x_vals = np.linspace(0, 4, 100)
y_disc = a_discriminative + b_discriminative * x_vals
y_gen = a_inverted + b_inverted * x_vals

# Plot
plt.scatter(x, y, label="Data")
plt.plot(x_vals, y_disc, label="Discriminative", color="blue")
plt.plot(x_vals, y_gen, label="Generative (inverted)", color="red", linestyle="--")
plt.legend()
plt.xlabel("x")
plt.ylabel("y")
plt.title("Discriminative vs. Generative Linear Fit")
plt.grid(True)
plt.show()
```

## Problem 6


We are given a shallow neural network:

### Hidden unit:
$$
h_1(x) = a(\theta_{10} + \theta_{11} x)
$$

### Output:
$$
y = \phi_0 + \phi_1 \cdot h_1(x)
$$

Now we apply the transformation:

- Multiply both $\theta_{10}$ and $\theta_{11}$ by a constant $\alpha$
- Divide $\phi_1$ by the same constant $\alpha$

---

### üîπ Case 1: Positive scaling ($\alpha > 0$)

We substitute:

$$
h_1(x) = a(\alpha \theta_{10} + \alpha \theta_{11} x) = a\left(\alpha (\theta_{10} + \theta_{11}x)\right)
$$

Using the ReLU property:  
If $a(z) = \max(0, z)$ and $\alpha > 0$, then:

$$
a(\alpha z) = \alpha \cdot a(z)
$$

So:

$$
h_1(x) = \alpha \cdot a(\theta_{10} + \theta_{11}x)
$$

Now the output becomes:

$$
y = \phi_0 + \frac{\phi_1}{\alpha} \cdot h_1(x) 
= \phi_0 + \frac{\phi_1}{\alpha} \cdot \alpha \cdot a(\theta_{10} + \theta_{11}x)
= \phi_0 + \phi_1 \cdot a(\theta_{10} + \theta_{11}x)
$$

‚úÖ **Conclusion:** If $\alpha > 0$, the scaling **cancels out**, and the network behaves **exactly the same**.

---

### üî∏ Case 2: Negative scaling ($\alpha < 0$)

Now we flip the sign of the input:

$$
h_1(x) = a(\alpha (\theta_{10} + \theta_{11}x))
$$

But since $\alpha < 0$, the argument inside the ReLU flips:

- If $\theta_{10} + \theta_{11}x > 0$, then $\alpha (\cdot) < 0$ ‚áí $a(\cdot) = 0$
- If $\theta_{10} + \theta_{11}x < 0$, then $\alpha (\cdot) > 0$ ‚áí $a(\cdot) > 0$

So:

- The ReLU now activates in the **opposite region**
- The output becomes a **different function**
- The shape is a **mirror** of the original

‚ùå **Conclusion:** If $\alpha < 0$, the network output **changes**, because ReLU is **not symmetric**.

---

### ‚úÖ Summary Table

| $\alpha$            | Effect on Output Function                            |
|---------------------|-------------------------------------------------------|
| $\alpha > 0$        | No change ‚Äî scaling cancels                          |
| $\alpha < 0$        | Function changes ‚Äî ReLU flips active region          |

**No**, the least squares loss for the network in Equation 3.1 **does not have a unique minimum**.

#### üîπ Reasons:

1. **Non-convex loss surface**:  
   Due to the nonlinear activation function (e.g. ReLU), the model is not convex in its parameters, so the loss surface can contain multiple local minima and saddle points.

2. **Parameter scaling invariance**:  
   For each hidden unit, scaling $\theta_{i0}, \theta_{i1}$ by $\alpha$ and dividing $\phi_i$ by $\alpha$ gives the same function.  
   So multiple parameter combinations lead to the **same output** and **same loss**.

3. **Hidden unit permutation symmetry**:  
   Swapping the roles of hidden units (e.g., exchanging $\phi_1$ with $\phi_2$ and their corresponding $\theta$'s) gives the same model output, so again, multiple parameter sets yield identical loss.

#### ‚úÖ Conclusion:
The loss function may have **multiple global minima** and **many equally good parameter sets**. There is not a single "best" set of parameters.

## Problem 8

We consider the network:

$$
y = \phi_0 + \phi_1 \cdot a(\theta_{10} + \theta_{11}x) + \phi_2 \cdot a(\theta_{20} + \theta_{21}x) + \phi_3 \cdot a(\theta_{30} + \theta_{31}x)
$$

With parameters:

- $\phi = \{-0.23, -1.3, 1.3, 0.66\}$
- $\theta = \{-0.2, 0.4, -0.9, 0.9, 1.1, -0.7\}$  
  (i.e., $\theta_{10}, \theta_{11}, \theta_{20}, \theta_{21}, \theta_{30}, \theta_{31}$)

We analyze the result of replacing ReLU with the following:

---

### üîπ 1. Heaviside Step Function

Defined as:

$$
\text{heaviside}(z) = 
\begin{cases}
0 & \text{if } z < 0 \\
1 & \text{if } z \geq 0
\end{cases}
$$

#### üî∏ Effect:
- Each hidden unit outputs either **0 or 1**
- Output $y$ becomes a **piecewise constant function**
- Up to 3 hidden units => up to **8 possible levels** (combinations of on/off)
- Transitions are sharp, **discontinuous jumps**

#### üî∏ Function Family:
- Step functions
- Useful for approximating classification boundaries or simple rules

---

### üîπ 2. Hyperbolic Tangent ($\tanh$)

Defined as:

$$
\tanh(z) = \frac{e^z - e^{-z}}{e^z + e^{-z}} \in (-1, 1)
$$

#### üî∏ Effect:
- Each unit outputs a **smooth curve** from -1 to 1
- The final output is a **smooth nonlinear function**
- No hard breakpoints ‚Äî only gradual bends

#### üî∏ Function Family:
- Smooth S-shaped functions
- Can represent gradual trends and non-linear dependencies

---

### üîπ 3. Rectangular Window Function

Defined as:

$$
\text{rect}(z) =
\begin{cases}
0 & z < 0 \\
1 & 0 \leq z \leq 1 \\
0 & z > 1
\end{cases}
$$

#### üî∏ Effect:
- Each unit produces a **flat bump** (1) in a specific input range
- Output is a **sum of flat-topped pulses**
- Can isolate contributions to narrow regions of $x$

#### üî∏ Function Family:
- Localized bumps or flat peaks
- Good for building **histogram-like** or spatially-aware functions

---

### ‚úÖ Summary Table

| Activation Function | Output Behavior             | Function Type                         |
|---------------------|-----------------------------|----------------------------------------|
| Heaviside           | Piecewise constant           | Step functions with jumps              |
| Tanh                | Smooth, continuous           | S-shaped nonlinear functions           |
| Rect                | Localized flat peaks         | Bumpy/histogram-like, spatially local  |

---

### üß† Informal Insight

- **ReLU** builds **piecewise linear** functions
- **Heaviside** builds **discrete step** functions
- **Tanh** builds **smooth, nonlinear** curves
- **Rect** builds **piecewise constant, localized bump** functions

Each activation expands or restricts the kinds of functions the network can express ‚Äî from sharp rule-like logic to smooth approximations or localized detection.



```python
import numpy as np
import matplotlib.pyplot as plt

# Input range
x = np.linspace(-1, 2, 500)

# Parameters from the problem
phi_0 = -0.23
phi_1 = -1.3
phi_2 = 1.3
phi_3 = 0.66

theta_10, theta_11 = -0.2, 0.4
theta_20, theta_21 = -0.9, 0.9
theta_30, theta_31 = 1.1, -0.7

# Pre-activations for each hidden unit
z1 = theta_10 + theta_11 * x
z2 = theta_20 + theta_21 * x
z3 = theta_30 + theta_31 * x

# Activation functions
def heaviside(z):
    return np.where(z >= 0, 1.0, 0.0)

def rect(z):
    return np.where((z >= 0) & (z <= 1), 1.0, 0.0)

def tanh(z):
    return np.tanh(z)

# Output functions
def output(a1, a2, a3):
    return phi_0 + phi_1 * a1 + phi_2 * a2 + phi_3 * a3

# Compute outputs
y_heaviside = output(heaviside(z1), heaviside(z2), heaviside(z3))
y_tanh = output(tanh(z1), tanh(z2), tanh(z3))
y_rect = output(rect(z1), rect(z2), rect(z3))

# Plotting
fig, axs = plt.subplots(3, 1, figsize=(8, 10), sharex=True)

axs[0].plot(x, y_heaviside, label="Heaviside", color="darkorange")
axs[0].set_title("Output with Heaviside Activation")
axs[0].set_ylabel("y")
axs[0].grid(True)

axs[1].plot(x, y_tanh, label="Tanh", color="teal")
axs[1].set_title("Output with Tanh Activation")
axs[1].set_ylabel("y")
axs[1].grid(True)

axs[2].plot(x, y_rect, label="Rect", color="purple")
axs[2].set_title("Output with Rect Activation")
axs[2].set_xlabel("x")
axs[2].set_ylabel("y")
axs[2].grid(True)

plt.tight_layout()
plt.show()
```

## Problem 10


A shallow neural network with 3 ReLU hidden units typically creates **4 linear regions**.  
However, fewer regions can occur under the following circumstances:

---

### üîπ 1. A hidden unit **never activates**
- If $\theta_{i0} + \theta_{i1}x < 0$ for all $x$ in the input domain, then:
  $$ h_i(x) = 0 \quad \Rightarrow \phi_i h_i(x) = 0 $$
- The unit has no effect, and its activation point doesn't create a region change.

---

### üîπ 2. A hidden unit is **always active**
- If $\theta_{i0} + \theta_{i1}x > 0$ for all $x$, then:
  $$ h_i(x) = \theta_{i0} + \theta_{i1}x $$
- This behaves like a linear term and doesn't introduce a kink.

---

### üîπ 3. Two units activate at the **same input**
- If two breakpoints overlap:
  $$ -\frac{\theta_{i0}}{\theta_{i1}} = -\frac{\theta_{j0}}{\theta_{j1}} $$
- Then two region transitions collapse into one.

---

### üîπ 4. The hidden unit's effect is **canceled in the output**
- If $\phi_i = 0$ or multiple units produce changes that cancel out, their activation doesn't create a visible change in $y(x)$.

---

### ‚úÖ Conclusion:

Although 3 ReLU units can theoretically create 4 linear regions, the network may produce fewer if:
- Units are always or never active
- Breakpoints overlap
- Output weights cancel effects

$$
L(\phi) = -\log Pr(y \mid f(\mathbf{x}, \phi), \kappa)
= -\kappa \cos(y - f(\mathbf{x}, \phi)) + \log(2\pi I_0(\kappa))
$$

Since $ \kappa $ is constant, the loss simplifies to:

$$
\boxed{L(\phi) = -\cos(y - f(\mathbf{x}, \phi))}
\quad \text{(ignoring constants)}
$$


$$
\cos(y - \mu)
$$

achieves its **maximum when** $ y = \mu $, the distribution is also **maximized** at:

$$
\boxed{y = \mu}
$$

So the **mode** (most likely value) of the von Mises distribution is exactly $ \mu $.


---

### ‚úÖ Inference

Once trained, the most likely prediction for $ y $ given $ \mathbf{x} $ is:

$$
\boxed{\hat{y} = f(\mathbf{x}, \phi)}
$$

This is the **mode** of the von Mises distribution.


## Problem 4


We model $ y \in \mathbb{R} $ given input $ x $ using a **mixture of two Gaussians**:

$$
Pr(y \mid x) = \lambda \cdot \mathcal{N}(y \mid \mu_1, \sigma_1^2) + (1 - \lambda) \cdot \mathcal{N}(y \mid \mu_2, \sigma_2^2)
$$

Here, $ \mathcal{N}(y \mid \mu, \sigma^2) $ refers to the **Gaussian (normal) probability density function**, defined as:

$$
\mathcal{N}(y \mid \mu, \sigma^2) = \frac{1}{\sqrt{2\pi \sigma^2}} \exp\left(-\frac{(y - \mu)^2}{2\sigma^2}\right)
$$

---

### ‚úÖ Loss Function

Use the **negative log-likelihood** over $ I $ training examples:

$$
\mathcal{L}(\phi) = -\sum_{i=1}^I \log\left[
\lambda_i \cdot \mathcal{N}(y_i \mid \mu_{1i}, \sigma_{1i}^2) + (1 - \lambda_i) \cdot \mathcal{N}(y_i \mid \mu_{2i}, \sigma_{2i}^2)
\right]
$$

where all parameters $ \lambda_i, \mu_{1i}, \mu_{2i}, \sigma_{1i}, \sigma_{2i} $ are predicted by the model $ f(x_i, \phi) $.

---

### ‚ùó Inference Challenges

- **Multimodality**: There may be two plausible outputs $ \mu_1, \mu_2 $ . Which one do you pick?
- **Ambiguity**: Choosing only the most likely mode may discard useful information.
- **Mode collapse**: The model might default to using just one component (e.g., $ \lambda \to 1 $).
- **Numerical instability**: Very small variances can cause sharp peaks, leading to unstable gradients.

---

### ‚úÖ Inference Strategy

Once trained, given a new $ x $, the model outputs:

$$
(\lambda, \mu_1, \mu_2, \sigma_1, \sigma_2)
$$

You may:
- Return the most likely mode:
  $$
  \hat{y} = \begin{cases}
  \mu_1 & \text{if } \lambda > 0.5 \\
  \mu_2 & \text{otherwise}
  \end{cases}
  $$
- Sample from the mixture distribution
- Return both modes (if downstream task can use them)

- $ \sigma^2 $ is constant (not learned)

---

The total likelihood is:

$$
Pr(\mathbf{y} \mid f(\mathbf{x}, \phi)) =
\prod_{d=1}^{10} \frac{1}{\sqrt{2\pi\sigma^2}} \exp\left( -\frac{(y_d - \mu_d)^2}{2\sigma^2} \right)
$$

---

### ‚úÖ Negative Log-Likelihood

$$
\mathcal{L}(\phi) = \sum_{d=1}^{10} \left[ \frac{(y_d - \mu_d)^2}{2\sigma^2} + \frac{1}{2} \log(2\pi\sigma^2) \right]
$$

$$
= \frac{1}{2\sigma^2} \sum_{d=1}^{10} (y_d - \mu_d)^2 + \text{const}
$$

---

### ‚úÖ Conclusion

Minimizing the negative log-likelihood is equivalent to minimizing:

$$
\boxed{ \sum_{d=1}^{10} (y_d - \mu_d)^2 }
$$

which is the **sum of squared errors**, when $ \sigma^2 $ is fixed.

### ‚ùó Problems Caused by Different Scales

1. **Loss imbalance**: Larger-magnitude targets (e.g., weight) dominate the loss, making the model biased toward predicting those better.
2. **Gradient imbalance**: Larger outputs produce larger gradients, potentially destabilizing or slowing training for smaller targets.

---

### ‚úÖ Solution 1: Normalize the Targets

Standardize each output to have zero mean and unit variance:

$$
\tilde{y}_d = \frac{y_d - \mu_d}{\sigma_d}
$$

Then invert this transformation after prediction.

### ‚úÖ Solution 2: Weighted Loss Function

Use weights in the loss function to balance target scales:

$$
\text{Loss} = \alpha \cdot (y_{\text{height}} - \hat{y}_{\text{height}})^2 + \beta \cdot (y_{\text{weight}} - \hat{y}_{\text{weight}})^2
$$

where $ \alpha, \beta $ are chosen based on inverse target variance or importance.

---

Both approaches ensure balanced training and fair optimization of all targets.

## Problem 2


We are given the loss function from Equation (6.5) for a simple linear regression model:

$$
L[\phi] = \sum_{i=1}^I \left( \phi_0 + \phi_1 x_i - y_i \right)^2
$$

We are asked to:

1. Derive the **Hessian matrix** $ \mathbf{H}[\phi] $
2. Prove that the function is **convex** by showing that:
   - Its **eigenvalues** are always positive  
   - This can be done by checking the **trace** and **determinant**

---

### ‚úÖ Step 1: Define the Residual

Let:

$$
r_i = \phi_0 + \phi_1 x_i - y_i
$$

Then:

$$
L[\phi] = \sum_{i=1}^I r_i^2
$$

---

### ‚úÖ Step 2: First Derivatives (Gradient)

$$
\frac{\partial L}{\partial \phi_0} = \sum_{i=1}^I 2r_i \cdot \frac{\partial r_i}{\partial \phi_0} = \sum_{i=1}^I 2(\phi_0 + \phi_1 x_i - y_i)
$$

$$
\frac{\partial L}{\partial \phi_1} = \sum_{i=1}^I 2r_i \cdot \frac{\partial r_i}{\partial \phi_1} = \sum_{i=1}^I 2x_i(\phi_0 + \phi_1 x_i - y_i)
$$

---

### ‚úÖ Step 3: Second Derivatives ‚Äî The Hessian Matrix

The **Hessian** is the matrix of all second partial derivatives:

$$
\mathbf{H}[\phi] =
\begin{bmatrix}
\frac{\partial^2 L}{\partial \phi_0^2} & \frac{\partial^2 L}{\partial \phi_0 \partial \phi_1} \\
\frac{\partial^2 L}{\partial \phi_1 \partial \phi_0} & \frac{\partial^2 L}{\partial \phi_1^2}
\end{bmatrix}
$$

Let‚Äôs compute each term:

---

#### üîπ $ \frac{\partial^2 L}{\partial \phi_0^2} $

We differentiate $ \frac{\partial L}{\partial \phi_0} = \sum 2(\phi_0 + \phi_1 x_i - y_i) $:

Only $ \phi_0 $ remains variable. Derivative is:

$$
\frac{\partial^2 L}{\partial \phi_0^2} = \sum_{i=1}^I 2 = 2I
$$

---

#### üîπ $ \frac{\partial^2 L}{\partial \phi_1^2} $

Differentiate $ \frac{\partial L}{\partial \phi_1} = \sum 2x_i(\phi_0 + \phi_1 x_i - y_i) $:

$$
\frac{\partial^2 L}{\partial \phi_1^2} = \sum_{i=1}^I 2x_i^2
$$

---

#### üîπ Mixed Partials: $ \frac{\partial^2 L}{\partial \phi_0 \partial \phi_1} $ and $ \frac{\partial^2 L}{\partial \phi_1 \partial \phi_0} $

From:

$$
\frac{\partial L}{\partial \phi_0} = \sum 2(\phi_0 + \phi_1 x_i - y_i)
$$

Differentiate w.r.t. $ \phi_1 $:

> ‚ùó Here‚Äôs the key logic:
> - When taking \( \frac{\partial}{\partial \phi_1} \), we **treat \( \phi_0 \) as constant only for that step**
> - We do **not remove** \( \phi_1 \); we treat everything carefully

$$
\frac{\partial^2 L}{\partial \phi_0 \partial \phi_1} = \sum_{i=1}^I 2x_i
$$

By symmetry:

$$
\frac{\partial^2 L}{\partial \phi_1 \partial \phi_0} = \sum_{i=1}^I 2x_i
$$

---

### ‚úÖ Final Hessian Matrix:

$$
\mathbf{H}[\phi] =
\begin{bmatrix}
2I & 2 \sum x_i \\
2 \sum x_i & 2 \sum x_i^2
\end{bmatrix}
$$

---

### ‚úÖ Step 4: Prove Convexity

A function is **convex** if its Hessian is **positive semidefinite** (all eigenvalues ‚â• 0). A 2√ó2 matrix is positive definite if:

- **Trace is positive**
- **Determinant is positive**

---

#### üîπ Trace:

$$
\text{Tr}(\mathbf{H}) = 2I + 2 \sum x_i^2 > 0
$$

---

#### üîπ Determinant:

$$
\det(\mathbf{H}) = (2I)(2 \sum x_i^2) - (2 \sum x_i)^2 = 4I \sum x_i^2 - 4(\sum x_i)^2
$$

$$
= 4 \left( I \sum x_i^2 - (\sum x_i)^2 \right)
$$

This is always **non-negative** by the **Cauchy‚ÄìSchwarz inequality**, and **positive unless all \( x_i \) are equal**.

---

### ‚úÖ Conclusion

Since the trace and determinant are positive, all eigenvalues of the Hessian are positive ‚áí  
The function $ L[\phi] $ is **convex** and has a **unique global minimum**.

## Problem 5


We are given the binary classification loss function:

\ell_i = -(1 - y_i) \log\left(1 - \text{sig}(f(x_i, \phi))\right) - y_i \log\left(\text{sig}(f(x_i, \phi))\right)

$$

Let:

- $ z = f(x_i, \phi) $
- $ \hat{y}_i = \text{sig}(z) $

Then the loss becomes:

$$
\ell_i = -(1 - y_i)\log(1 - \hat{y}_i) - y_i\log(\hat{y}_i)
$$

Now compute:

$$
\frac{\partial \ell_i}{\partial z}
= -(1 - y_i) \cdot \left(\frac{-1}{1 - \hat{y}_i} \cdot \frac{d\hat{y}_i}{dz} \right)
  - y_i \cdot \left(\frac{1}{\hat{y}_i} \cdot \frac{d\hat{y}_i}{dz} \right)
$$

The derivative of sigmoid is:

$$
\frac{d\hat{y}_i}{dz} = \hat{y}_i(1 - \hat{y}_i)
$$

Substitute:

$$
\frac{\partial \ell_i}{\partial z}
= \left( \frac{1 - y_i}{1 - \hat{y}_i} - \frac{y_i}{\hat{y}_i} \right) \cdot \hat{y}_i(1 - \hat{y}_i)
$$

Simplify:

$$
\frac{\partial \ell_i}{\partial z}
= \hat{y}_i - y_i
$$

---

### ‚úÖ Final Answer:

$$
\boxed{
\frac{\partial \ell_i}{\partial f(x_i, \phi)} = \text{sig}(f(x_i, \phi)) - y_i
}
$$

This is the classic result used in logistic regression gradient computations.


## Problem 6


We are given:

$$
\mathbf{z} = \boldsymbol{\beta} + \Omega \mathbf{h}
$$

We are asked to compute the matrix of partial derivatives:

$$
\frac{\partial \mathbf{z}}{\partial \mathbf{h}}
$$

---

### üîπ Step 1: Write element-wise definition

Let $ \mathbf{z} \in \mathbb{R}^m $, $ \mathbf{h} \in \mathbb{R}^n $, and $ \Omega \in \mathbb{R}^{m \times n} $. Then:

$$
z_i = \beta_i + \sum_{j=1}^n \Omega_{ij} h_j
$$

---

### üîπ Step 2: Compute partial derivative

We compute each entry of the Jacobian matrix $ \frac{\partial z_i}{\partial h_j} $:

$$
\frac{\partial z_i}{\partial h_j} = \Omega_{ij}
$$

So the $ (i, j) $-th entry of the Jacobian is $ \Omega_{ij} $.

---

### üîπ Step 3: Matrix Form

This means the Jacobian $ \frac{\partial \mathbf{z}}{\partial \mathbf{h}} $ has:

- rows indexed by $ i $
- columns indexed by $ j $

and value $ \Omega_{ij} $, i.e.:

$$
\frac{\partial \mathbf{z}}{\partial \mathbf{h}} = \Omega
$$

But note: by convention, the Jacobian is **$ [\partial z_i / \partial h_j] $**, meaning **row = output index**, **column = input index**. So to match matrix multiplication rules (chain rule):

$$
\boxed{
\frac{\partial \mathbf{z}}{\partial \mathbf{h}} = \Omega^T
}
$$

---

### ‚úÖ Final Result

$$
\boxed{
\frac{\partial \mathbf{z}}{\partial \mathbf{h}} = \Omega^T

This is a **linear regression** problem in terms of $ \beta, \omega_1, \omega_2, \omega_3 $, with a least squares loss:

$$
L = \sum_{i=1}^I \left( y_i - \left[\beta + \omega_1 h_1(x_i) + \omega_2 h_2(x_i) + \omega_3 h_3(x_i)\right] \right)^2
$$

---

### üßÆ Step 1: Linear Regression Form

Let‚Äôs rewrite this in matrix form:

- Let $ H \in \mathbb{R}^{I \times 4} $ be the design matrix:
$$
H_i = [1, h_1(x_i), h_2(x_i), h_3(x_i)]
$$

- Let $ \theta = \begin{bmatrix} \beta \\ \omega_1 \\ \omega_2 \\ \omega_3 \end{bmatrix} \in \mathbb{R}^4 $

- Let $ Y = \begin{bmatrix} y_1 \\ y_2 \\ \vdots \\ y_I \end{bmatrix} \in \mathbb{R}^I $

Then the model becomes:

$$
Y = H \theta + \varepsilon
$$

---

### üß† Step 2: Least Squares Solution

The closed-form solution that minimizes the squared error is:

$$
\theta = (H^T H)^{-1} H^T Y
$$

This gives the optimal values of:

- $ \beta $
- $ \omega_1 $
- $ \omega_2 $
- $ \omega_3 $

---

### ‚úÖ Final Answer

To find $ \beta $, $ \omega_1 $, $ \omega_2 $, and $ \omega_3 $ in closed form:

1. Compute the hidden layer activations for all inputs: $ h_1(x_i), h_2(x_i), h_3(x_i) $
2. Construct matrix $ H \in \mathbb{R}^{I \times 4} $ with rows $ [1, h_1(x_i), h_2(x_i), h_3(x_i)] $
3. Use the least squares formula:

$$
\boxed{
\theta = (H^T H)^{-1} H^T Y
}
$$

This solves for the optimal output layer parameters using standard linear regression.

### ‚ùì Question Recap:
We are considering the **overparameterized** regime:
- The model has **more capacity** than training data.
- It can reduce the **training loss to zero**.
- The model is **heteroscedastic** (it predicts both the mean and the variance of the output).

---

### ‚ö†Ô∏è Implications for Heteroscedastic Models

In a **heteroscedastic model**, the network predicts:
- A **mean** output: $\mu(x)$
- A **variance** (or noise level): $\sigma^2(x)$

The loss is typically based on the **negative log-likelihood** of a Gaussian:

$$
\mathcal{L}(x, y) = \frac{(y - \mu(x))^2}{2\sigma^2(x)} + \frac{1}{2} \log \sigma^2(x)
$$

---

### üö® Problem: When the model can fit the training data **perfectly**
- It can make the **first term** (the squared error) go to **zero**.
- This incentivizes the model to set **$\sigma^2(x) \to 0$**, since it further reduces the loss.
- But this leads to **numerical instability** or **overconfident predictions** (variance approaching zero).
- The model **overfits**, especially on noisy data.

---

### ‚úÖ Proposed Solution: Variance Regularization

To resolve this, we can:
1. **Regularize** the predicted variance:
   - Add a penalty term like:  
     $$
     \lambda \cdot \mathbb{E}_x \left[ \log \sigma^2(x) \right]
     $$
   - This prevents the model from shrinking variance too aggressively.

2. **Use a prior** (Bayesian approach):
   - Place a prior over $\sigma^2(x)$ or use variational inference.
   - This keeps predictions calibrated, especially in low-data regions.

3. **Cap the variance floor**:
   - Add a small constant $\epsilon$ to prevent division by near-zero values:
     $$
     \frac{1}{\sigma^2(x) + \epsilon}
     $$

---

### üìå Summary

| Issue                         | Consequence                                | Solution                          |
|------------------------------|---------------------------------------------|-----------------------------------|
| $\mathcal{L}$ allows $\sigma^2(x) \to 0$ when error = 0 | Overconfident, unstable variance estimates | Regularize or constrain $\sigma^2(x)$ |

Controlling predictive variance is critical in overparameterized heteroscedastic models to ensure generalization and stability.

Show that two random points drawn from a **1000-dimensional standard Gaussian distribution**

---

### üß† Intuition

Let:
- $\mathbf{x}, \mathbf{y} \in \mathbb{R}^{1000}$ be two vectors where each component is drawn independently from $\mathcal{N}(0, 1)$.

We want to show that:
- The **angle** $\theta$ between $\mathbf{x}$ and $\mathbf{y}$ tends to **$\frac{\pi}{2}$** as dimensionality increases.
- That is, their **dot product** is close to zero.

---

### üìê Step 1: Normalize the Vectors

Let:
- $u = \frac{\mathbf{x}}{\|\mathbf{x}\|}$  
- $v = \frac{\mathbf{y}}{\|\mathbf{y}\|}$  

Then the **cosine of the angle** between them is:

$
\cos\theta = u \cdot v = \frac{\mathbf{x} \cdot \mathbf{y}}{\|\mathbf{x}\| \cdot \|\mathbf{y}\|}
$

---

### üìä Step 2: Analyze Expected Value of Dot Product

- Each entry of $\mathbf{x}$ and $\mathbf{y}$ is independent with mean 0 and variance 1.
- So the dot product $\mathbf{x} \cdot \mathbf{y} = \sum_{i=1}^{d} x_i y_i$ is the sum of $d$ i.i.d. terms with:
  - $\mathbb{E}[x_i y_i] = 0$  
  - $\text{Var}[x_i y_i] = 1$

By **Central Limit Theorem**:

$
\mathbf{x} \cdot \mathbf{y} \sim \mathcal{N}(0, d)
$

---

### üìê Step 3: Analyze Norms

Each vector has:

$
\|\mathbf{x}\|^2 = \sum_{i=1}^{d} x_i^2 \sim \chi^2_d
$

So:

- $\mathbb{E}[\|\mathbf{x}\|^2] = d$
- $\|\mathbf{x}\| \approx \sqrt{d}$

Same for $\|\mathbf{y}\|$

---

### üìâ Step 4: Dot Product Scales Like $O(\sqrt{d})$, Norm Scales Like $O(\sqrt{d})$

So:

$
\cos\theta = \frac{\mathbf{x} \cdot \mathbf{y}}{\|\mathbf{x}\| \|\mathbf{y}\|} \sim \frac{\mathcal{N}(0, d)}{d} \to 0 \quad \text{as } d \to \infty
$

Therefore:

- $\theta \to \frac{\pi}{2}$ (vectors become orthogonal)

---

### ‚úÖ Conclusion

In high-dimensional space, **two random vectors from a standard Gaussian** are **almost orthogonal** with high probability. This is due to the concentration of measure and the law of large numbers.

This property is fundamental in high-dimensional geometry and is one reason why many ML techniques (e.g., random projections, kernels) work well in high dimensions.

&+ f_3[x + f_1[x] + f_2[x + f_1[x]]] \\
&+ f_4[x + f_1[x] + f_2[x + f_1[x]] + f_3[x + f_1[x] + f_2[x + f_1[x]]]]
\end{aligned}
$$

Now we compute $ \frac{\partial y}{\partial f_1} $. Note that $ f_1 $ appears in all terms (either directly or nested), so we apply the chain rule repeatedly:

$$
\frac{\partial y}{\partial f_1} =
\underbrace{\mathbf{I}}_{\text{from } f_1[x]} +
\underbrace{\frac{\partial f_2}{\partial f_1}}_{\text{via } f_2[x + f_1[x]]} +
\underbrace{\left( \frac{\partial f_3}{\partial f_1} + \frac{\partial f_2}{\partial f_1} \frac{\partial f_3}{\partial f_2} \right)}_{\text{chain rule into } f_3[\cdots]} +
\underbrace{\left( 
\frac{\partial f_4}{\partial f_1}
+ \frac{\partial f_2}{\partial f_1} \frac{\partial f_4}{\partial f_2}
+ \frac{\partial f_3}{\partial f_1} \frac{\partial f_4}{\partial f_3}
+ \frac{\partial f_2}{\partial f_1} \frac{\partial f_3}{\partial f_2} \frac{\partial f_4}{\partial f_3}
\right)}_{\text{all paths into } f_4}
$$

This matches equation (11.6):

$$
\frac{\partial y}{\partial f_1} = \mathbf{I}
+ \frac{\partial f_2}{\partial f_1}
+ \left( \frac{\partial f_3}{\partial f_1} + \frac{\partial f_2}{\partial f_1} \frac{\partial f_3}{\partial f_2} \right)
+ \left(
\frac{\partial f_4}{\partial f_1}
+ \frac{\partial f_2}{\partial f_1} \frac{\partial f_4}{\partial f_2}
+ \frac{\partial f_3}{\partial f_1} \frac{\partial f_4}{\partial f_3}
+ \frac{\partial f_2}{\partial f_1} \frac{\partial f_3}{\partial f_2} \frac{\partial f_4}{\partial f_3}
\right)
$$

---

### üí¨ What is **ùêà**?

The **ùë∞** represents the **identity matrix** (or identity operator). This comes from:

$$
\frac{\partial f_1[x]}{\partial f_1[x]} = \mathbf{I}
$$

It means that $ f_1[x] $ appears directly in the output $ y $, so any change in $ f_1 $ changes $ y $ identically in that term.

---

### ‚úÖ Conclusion

We showed that each higher-order function $ f_2, f_3, f_4 $ depends on $ f_1[x] $ through nested applications of the chain rule, and the total derivative accumulates all paths that start with $ f_1 $, leading to equation (11.6).

### üîç Context

In **Figure 11.6a**, each residual block has:
- A main branch that applies a function \( f_i \)
- A shortcut or skip connection that directly passes the input \( x \)

The output is the sum of these two branches:
$$
z = x + f_i(x)
$$

---

### ‚ùì Why Are the Branches Uncorrelated?

At initialization:
- $ x $ is the input.
- $ f_i(x) $ applies a nonlinear transformation to $ x $ using random weights.
- Modern initialization schemes (e.g. He or Xavier) ensure that:
  - $ \mathbb{E}[f_i(x)] = 0 $
  - $ x $ and $ f_i(x) $ are approximately **uncorrelated**.

So:
> The skip connection and the transformed output are uncorrelated random variables.

---

### üìê Variance of Sum of Uncorrelated Variables

Let:
- $ z = x + f(x) $
- $ x $ and $ f(x) $ are uncorrelated
- Both have zero mean

Then:

$$
\text{Var}[z] = \text{Var}[x + f(x)] = \text{Var}[x] + \text{Var}[f(x)] + 2 \cdot \text{Cov}(x, f(x))
$$

Since $ \text{Cov}(x, f(x)) = 0 $, this simplifies to:

$$
\boxed{
\text{Var}[x + f(x)] = \text{Var}[x] + \text{Var}[f(x)]
}
$$

---

### ‚úÖ Conclusion

- The values in the two branches are uncorrelated due to random initialization and architectural separation.
- The variance of the sum of uncorrelated variables is the **sum of their variances**.
